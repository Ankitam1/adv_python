# Dunder functions
1. \_\_repr\_\_ -> proprer representation for class
2. \_\_str\_\_ -> print representation for class

```python
class Person2:
    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def __repr__(self):
        return (
            f"Person(name = {self.name},"
            f"age={self.age}, gender={self.gender}, occupation={self.occupation})"
        )

    def __str__(self):
        return f"name:{self.name}, age:{self.age}, gender:{self.gender}, occ:{self.occupation}"

    def intro(self):
        print(f"My name is {self.name}")
        print(f"I am {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as a {self.occupation}")
```

# Create a class to perform banking related transactions


```python
from datetime import datetime

datetime.now()
```

```python
datetime.now()
```

```python
print(datetime.now())
```

```python
from dataclasses import field
```

```python
@dataclass
class Transaction:
    amount: float
    timestamp: datetime = field(default_factory=datetime.now)
    description: str = ""
```

```python
import uuid

uuid.uuid4()
```

```python
str(uuid.uuid4())
```

```python
@dataclass
class BankAccount:
    owner: str
    balance: float = field(default=0.0, repr=False)
    account_id: str = field(default_factory=lambda: str(uuid.uuid4()))
    transactions: list[Transaction] = field(default_factory=list, repr=False)

    def deposit(self, amt: float, description: str = "deposit"):
        if amt <= 0:
            raise ValueError("Amount cannot be zero or negative")
        self.balance = self.balance + amt
        self.transactions.append(Transaction(amount=amt, description=description))

    def withdraw(self, amt: float, description: str = "withdraw"):
        if amt <= 0:
            raise ValueError("Amount cannot be zero or negative")
        if self.balance < amt:
            raise ValueError("Insufficient Balance")
        self.balance = self.balance - amt
        self.transactions.append(Transaction(amount=amt, description=description))

    def transfer(
        self, amt: float, target_account: "BankAccount", description="transfer"
    ):
        self.withdraw(amt, description=description)
        target_account.deposit(amt, description=description)

    def get_statement(self):
        lines = [f"Bank Statement for {self.owner} (Account no - {self.account_id})"]
        for tx in self.transactions:
            lines.append(f"{tx.timestamp} | {tx.description} | {tx.amount:.2f} INR")
        lines.append(f"Current Balance : {self.balance:.2f} INR")
        return "\n".join(lines)
```
